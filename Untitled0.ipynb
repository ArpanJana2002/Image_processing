Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/cat-7094808_1280.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('resized_cat.jpg')
```

```python

img_resized = mpimg.imread('/content/resized_cat.jpg')
img_res_plot = plt.imshow(img_resized)
plt.show()
```

```python
print(img_resized.shape)
```

Converting RGB images to Grayscale image using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/cat-7094808_1280.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
